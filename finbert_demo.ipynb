```python
from transformers import pipeline
```

```python
# Chargement du modèle FinBERT
classifier = pipeline("sentiment-analysis", model="ProsusAI/finbert")
```

Le classifier est une fonction qui prend une phrase en entrée et renvoie une liste contenant la prédiction, avec un label et un score de confiance

```python
print("Tape une phrase financière en anglais. Tape 'quit' pour arrêter.\n")
while True:
    text = input("Phrase : ")

    if text.lower() == "quit":
        print("Fin de la démo.")
        break

    result = classifier(text)[0]

    label = result["label"]
    score = result["score"]

    if label.lower() == "positive":
        emoji = "🟢"
    elif label.lower() == "negative":
        emoji = "🔴"
    else:
        emoji = "⚪"

    print(f"→ {emoji} Sentiment prédit : {label} | Confiance : {round(result['score']*100, 2)}%\n")
```
