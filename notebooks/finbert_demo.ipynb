```python
from transformers import pipeline
```

```python
# Chargement du modèle FinBERT
classifier = pipeline("sentiment-analysis", model="ProsusAI/finbert")
```

Le classifier est une fonction qui prend une phrase en entrée et renvoie une liste contenant la prédiction, avec un label et un score de confiance

```python
# Demande à l'utilisateur d'entrer une phrase
print("Tape une phrase financière en anglais. Tape 'quit' pour arrêter.\n")

# Boucle infinie pour tester plusieurs phrases
while True:
    text = input("Phrase : ")

    # Condition pour arrêter la démo
    if text.lower() == "quit":
        print("Fin de la démo.")
        break

    # Prédiction du modèle (FinBERT)
    result = classifier(text)[0]

    # Récupération du label et du score
    label = result["label"]
    score = result["score"]

    # Association d’un emoji selon le sentiment
    if label.lower() == "positive":
        emoji = "🟢"
    elif label.lower() == "negative":
        emoji = "🔴"
    else:
        emoji = "⚪"

    # Affichage du résultat
    print(f"→ {emoji} Sentiment prédit : {label} | Confiance : {round(score*100, 2)}%\n")
```
